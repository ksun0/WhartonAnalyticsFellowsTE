```python
import pandas as pd
```

```python
#  read individual datasets
dist_sales = pd.read_csv("distributor_sales.csv",index_col=0)
cma = pd.read_csv("cma.csv",index_col=0)
direct_sales = pd.read_csv("direct_sales.csv",index_col=0)
part = pd.read_csv("part.csv",index_col=0)
dist_inv = pd.read_csv("distributor_inventory.csv",index_col=0)
customer_df = pd.read_csv("new_customer.csv",index_col=0)
```

```python
# join parts and dist sales
dist_sales_part_joined_df = dist_sales.merge(
    part, how='inner', on=['te_corporate_part_number'])
```

```python
# join customer
distributor_sales_part_customer_df_joined = dist_sales_part_joined_df.merge(
    customer_df, how='inner',  left_on=['distributor_customer_id'], right_on=['customer_id'])
```

```python
# drop selected variables
drop_ls = ['part_material_type_name', 
           'part_engineering_status_name', 
           'part_discontinuance_status_name',
           'part_base_cpr_unit_of_measure_label', 
           'product_owning_business_unit_group_name', 
           'product_owning_profit_center_group_name', 
           'customer_id', 
           'fiscal_month_end_date_historical',
           'fiscal_month_begin_date_converted_historical',
           'fiscal_month_week_count', 
           'customer_type_description']
```

```python
new_master = distributor_sales_part_customer_df_joined.drop(columns=drop_ls)
```

```python
# select america and appliance as 1st focus
us_apl = new_master[(new_master['product_owning_business_unit_name'] == 'Appliances') & 
                    (new_master['customer_region'] == 'Americas')]
```

```python
us_apl.shape
```

```python
# train test split; 2015-2018 Q1/Q2 for training, 2018 Q3/4 + 2019 for testing
# training
us_apl_train_2015 = us_apl[us_apl['fiscal_year_historical'] == 2015]
us_apl_train_2016 = us_apl[us_apl['fiscal_year_historical'] == 2016]
us_apl_train_2017 = us_apl[us_apl['fiscal_year_historical'] == 2017]
us_apl_train_2018 = us_apl[(us_apl['fiscal_year_historical'] == 2018) & 
                    (us_apl['fiscal_quarter_historical'].isin([1,2]))]
```

```python
us_apl_train = pd.concat([us_apl_train_2015,
                          us_apl_train_2016,
                          us_apl_train_2017,
                          us_apl_train_2018],ignore_index=True)
```

```python
# test data
us_apl_test_2019 = us_apl[us_apl['fiscal_year_historical'] == 2019]
us_apl_test_2018 = us_apl[(us_apl['fiscal_year_historical'] == 2018) & 
                    (us_apl['fiscal_quarter_historical'].isin([3,4]))]
```

```python
us_apl_test = pd.concat([us_apl_test_2018,us_apl_test_2019],ignore_index=True)
```
